```python
import pyautogui
import pyperclip
import time

pyautogui.PAUSE = 1
 
# Passo 1: entrar no sitema da empresa (link no drive)
pyautogui.hotkey('ctrl', 't')
pyperclip.copy('https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga')
pyautogui.hotkey("ctrl","v")
pyautogui.press('enter')
time.sleep(8)

# Passo 2: Navegar no sitema e encontrar a Base de vendas
pyautogui.click(x=403, y=265, clicks=2)
time.sleep(3)

# Passo 3: Fazer o download da base de vendas
pyautogui.click(x=383, y=461)
pyautogui.click(x=1401, y=154)
pyautogui.click(x=1230, y=558)
```

```python
# passo 4: importar a base de vendas para python
import pandas as pd

tabela = pd.read_excel(r"C:\Users\doa\Downloads\Vendas - Dez.xlsx")
display(tabela)
```

```python
# Passo 5: Calcular os indicadores
faturamento = tabela["Valor Final"].sum()
print(faturamento)
quantidade = tabela["Quantidade"].sum()
print(quantidade)
```

```python
# Passo 6: enviar un email para a directoria com os indicadores de vendas

# abrir aba
pyautogui.hotkey("ctrl", "t")
time.sleep(3)
# entrar no email
pyperclip.copy("https://mail.google.com/mail/u/0/#inbox")
pyautogui.hotkey("ctrl","v")
pyautogui.press('enter')
time.sleep(8)

# clicar no botao escrever
pyautogui.click(x=81, y=173) 
time.sleep(8)

# preencher as informacoes de email
pyautogui.write("jose.gregorio.hernadez0424@gmail.com")
time.sleep(3)
pyautogui.press("tab") # para seleccionar o email

pyautogui.press("tab") # para pular
pyautogui.write("relatorio de vendas")

pyautogui.press("tab")
texto = f"""
Prezados clientes
o relatorio da vendas 

ficou da siguiente forma
R${faturamento:,.2f}
{quantidade:,}
"""
pyautogui.write(texto)

# enviar email
pyautogui.hotkey("ctrl","enter")
```
